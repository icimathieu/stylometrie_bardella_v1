## Calculer la duree totale du corpus
Installe la dep `yt-dlp` si besoin (`pip install yt-dlp`). Execute la cellule suivante pour recuperer la duree de chaque URL puis la somme en hh:mm:ss.


```python
from datetime import timedelta
from yt_dlp import YoutubeDL

def get_video_duration(url: str, ydl: YoutubeDL) -> int | None:
    try:
        info = ydl.extract_info(url, download=False)
    except Exception as exc:
        print(f"Echec pour {url}: {exc}")
        return None
    duration = info.get("duration")
    if duration is None:
        print(f"Aucune duree renvoyee pour {url}")
        return None
    return int(duration)

ydl_opts = {
    "quiet": True,
    "skip_download": True,
    "nocheckcertificate": True,
}

durations = {}
total_seconds = 0

with YoutubeDL(ydl_opts) as ydl:
    for url in liste_urls_bardella:
        duration = get_video_duration(url, ydl)
        if duration is None:
            continue
        durations[url] = duration
        total_seconds += duration

print(f"Videos OK: {len(durations)}/{len(liste_urls_bardella)}")
print(f"Total seconds: {total_seconds}")
print(f"Total hh:mm:ss: {timedelta(seconds=total_seconds)}")
```
