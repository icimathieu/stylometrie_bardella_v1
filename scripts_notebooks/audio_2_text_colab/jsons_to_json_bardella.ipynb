```python
import os
import glob
import json

TRANSCRIPTS_DIR = "/content/drive/MyDrive/fichiers_colab/histosef/bardella/whisper_transcripts"
output_merged = "/content/drive/MyDrive/fichiers_colab/histosef/bardella/whisper_transcriptions_merged.json"

json_files = sorted(glob.glob(os.path.join(TRANSCRIPTS_DIR, "*.json")))
print("Nombre de fichiers JSON trouvés :", len(json_files))

all_texts = []          # si tu veux une liste de chaînes de caractères
all_objects = []        # si tu veux garder les audio_id en plus

for path in json_files:
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # liste de chaînes
    all_texts.append(data["text"])
    # liste d'objets (id + texte)
    all_objects.append(data)

# Choisis ce que tu veux sauver : all_texts OU all_objects
with open(output_merged, "w", encoding="utf-8") as f:
    json.dump(all_texts, f, ensure_ascii=False, indent=2)

print("Fusion terminée. Fichier écrit :", output_merged)
```
