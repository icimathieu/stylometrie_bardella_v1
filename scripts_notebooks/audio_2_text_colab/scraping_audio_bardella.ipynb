```python
import yt_dlp

ydl_opts = {
    "format": "bestaudio/best", "cookiefile": "/content/youtube_cookies_2.txt","extractor_args": {
    "youtube": {"player_client": ["default"]}
},  "outtmpl": os.path.join(AUDIO_DIR, "%(id)s.%(ext)s"),
    "postprocessors": [
        {
            "key": "FFmpegExtractAudio",
            "preferredcodec": "mp3",
            "preferredquality": "128",
        }
    ],
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download(urls)
```
